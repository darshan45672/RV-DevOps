# 1️⃣ Basic Podman Operations

## 🚀 Essential Commands Quick Reference

| Command | Purpose | Example |
|---------|---------|---------|
| `podman search` | Search for images | `podman search nginx` |
| `podman pull` | Download image | `podman pull nginx` |
| `podman run` | Create and run container | `podman run -d nginx` |
| `podman ps` | List containers | `podman ps -a` |
| `podman images` | List images | `podman images` |

| `podman rm` | Remove container | `podman rm mycontainer` |
| `podman rmi` | Remove image | `podman rmi nginx` |

## 🎯 Your First Container

Let's break down the most common Podman workflow:

````python
# Demo: Basic Podman Workflow Simulation
import time
import random

def simulate_podman_command(command, description):
    """Simulate a Podman command execution"""
    print(f"🔄 Executing: {command}")
    print(f"📝 Purpose: {description}")
    
    # Simulate command execution time
    time.sleep(random.uniform(0.5, 1.5))
    
    if "search" in command:
        print("📊 Results:")
        print("NAME                      DESCRIPTION                       STARS")
        print("docker.io/library/nginx   Official build of Nginx.          19000")
        print("docker.io/bitnami/nginx   Bitnami Nginx Docker Image        150")
    
    elif "pull" in command:
        print("⬇️  Downloading layers...")
        print("✅ Image pulled successfully!")
    
    elif "run" in command:
        container_id = f"{''.join(random.choices('0123456789abcdef', k=12))}"
        print(f"🚀 Container started with ID: {container_id}")
        if "-d" in command:
            print("🔄 Running in background (detached mode)")
        return container_id
    
    elif "ps" in command:
        print("📋 Container List:")
        print("CONTAINER ID  IMAGE   COMMAND     STATUS    PORTS     NAMES")
        print("a1b2c3d4e5f6  nginx   nginx -g... Up 2 min  80/tcp    my-web")
    
    elif "stop" in command:
        print("⏹️  Container stopped successfully!")
    
    elif "rm" in command:
        print("🗑️  Container removed successfully!")
        
    print("✅ Command completed!\n" + "="*50 + "\n")

# Simulate the basic workflow
print("🎬 DEMO: Complete Podman Workflow")
print("="*60)

# Step 1: Search for image
simulate_podman_command("podman search nginx", "Find available nginx images")

# Step 2: Pull image
simulate_podman_command("podman pull nginx", "Download nginx image locally")

# Step 3: Run container
container_id = simulate_podman_command("podman run -d --name my-web -p 8080:80 nginx", 
                                     "Create and run nginx container")

# Step 4: List running containers
simulate_podman_command("podman ps", "Show running containers")

# Step 5: Stop container
simulate_podman_command("podman stop my-web", "Stop the running container")

# Step 6: Remove container
simulate_podman_command("podman rm my-web", "Clean up - remove stopped container")

print("🎉 Complete workflow demonstrated!")
print("💡 Try these commands in your terminal to see the real output!")
````

# 2️⃣ Container Management

## 🔧 Advanced Container Operations

### Port Mapping & Environment Variables

Understanding how to properly configure containers for real applications:

```bash
# Port mapping examples
podman run -d -p 8080:80 nginx                    # Map host:container ports
podman run -d -p 127.0.0.1:8080:80 nginx         # Bind to specific IP
podman run -d -p 8080:80 -p 8443:443 nginx       # Multiple ports

# Environment variables
podman run -d -e NODE_ENV=production myapp        # Single variable
podman run -d --env-file ./config.env myapp       # From file
```

### Resource Management

```bash
# Memory and CPU limits
podman run -d --memory=512m --cpus=1.5 myapp      # Limit resources
podman run -d --restart=always myapp               # Auto-restart policy
```

Let's see this in action:

````python
# Interactive Container Configuration Builder
import json

def build_podman_command():
    """Interactive tool to build Podman run commands"""
    print("🛠️  Interactive Container Configuration Builder")
    print("="*50)
    
    # Base configuration
    config = {
        "image": "nginx",
        "name": "my-app",
        "detached": True,
        "ports": [],
        "environment": [],
        "volumes": [],
        "memory": None,
        "cpus": None,
        "restart": None
    }
    
    # Common configurations
    scenarios = {
        "1": {
            "name": "Web Server",
            "config": {
                "image": "nginx",
                "name": "web-server",
                "ports": ["8080:80"],
                "environment": ["NGINX_HOST=localhost"],
                "restart": "always"
            }
        },
        "2": {
            "name": "Database",
            "config": {
                "image": "postgres",
                "name": "database",
                "ports": ["5432:5432"],
                "environment": ["POSTGRES_PASSWORD=secret", "POSTGRES_DB=myapp"],
                "volumes": ["db-data:/var/lib/postgresql/data"],
                "memory": "1g"
            }
        },
        "3": {
            "name": "Python Development",
            "config": {
                "image": "python:3.11",
                "name": "python-dev",
                "ports": ["8000:8000"],
                "volumes": ["$(pwd):/workspace"],
                "environment": ["PYTHONPATH=/workspace"],
                "interactive": True
            }
        }
    }
    
    print("📋 Choose a scenario:")
    for key, value in scenarios.items():
        print(f"  {key}. {value['name']}")
    
    choice = "1"  # Default for demo
    selected = scenarios[choice]
    
    print(f"\n✅ Selected: {selected['name']}")
    print(f"📝 Configuration:")
    
    # Build command
    cmd_parts = ["podman run"]
    
    if selected['config'].get('detached', True):
        cmd_parts.append("-d")
    
    if 'interactive' in selected['config']:
        cmd_parts.append("-it")
    
    if 'name' in selected['config']:
        cmd_parts.append(f"--name {selected['config']['name']}")
    
    for port in selected['config'].get('ports', []):
        cmd_parts.append(f"-p {port}")
    
    for env in selected['config'].get('environment', []):
        cmd_parts.append(f"-e {env}")
    
    for vol in selected['config'].get('volumes', []):
        cmd_parts.append(f"-v {vol}")
    
    if selected['config'].get('memory'):
        cmd_parts.append(f"--memory={selected['config']['memory']}")
    
    if selected['config'].get('restart'):
        cmd_parts.append(f"--restart={selected['config']['restart']}")
    
    cmd_parts.append(selected['config']['image'])
    
    command = " \\\n  ".join(cmd_parts)
    
    print(f"\n🚀 Generated Command:")
    print("```bash")
    print(command)
    print("```")
    
    print(f"\n📊 Configuration Summary:")
    print(json.dumps(selected['config'], indent=2))
    
    print(f"\n💡 Usage Tips:")
    print(f"- This creates a {selected['name'].lower()} container")
    print(f"- Accessible via configured ports")
    print(f"- Environment variables are pre-configured")
    print(f"- Data persistence handled via volumes")

# Run the builder
build_podman_command()
````

````python
# Python Container Development Scenarios

def generate_python_script():
    """Generate sample Python scripts for container testing"""
    
    # Calculator script
    calculator_script = '''#!/usr/bin/env python3
import sys
import platform

def calculator():
    print("🔢 Container Calculator")
    print(f"Python Version: {platform.python_version()}")
    print(f"Platform: {platform.platform()}")
    print("-" * 30)
    
    if len(sys.argv) != 4:
        print("Usage: python calculator.py <num1> <operation> <num2>")
        print("Operations: +, -, *, /")
        return
    
    try:
        num1 = float(sys.argv[1])
        op = sys.argv[2]
        num2 = float(sys.argv[3])
        
        if op == '+':
            result = num1 + num2
        elif op == '-':
            result = num1 - num2
        elif op == '*':
            result = num1 * num2
        elif op == '/':
            result = num1 / num2 if num2 != 0 else "Error: Division by zero"
        else:
            result = "Error: Invalid operation"
        
        print(f"{num1} {op} {num2} = {result}")
        
    except ValueError:
        print("Error: Invalid numbers provided")

if __name__ == "__main__":
    calculator()
'''

    # Web API script
    api_script = '''#!/usr/bin/env python3
from http.server import HTTPServer, BaseHTTPRequestHandler
import json
import datetime
import platform

class APIHandler(BaseHTTPRequestHandler):
    def do_GET(self):
        if self.path == '/':
            self.send_response(200)
            self.send_header('Content-type', 'application/json')
            self.end_headers()
            
            response = {
                'message': 'Hello from Python container!',
                'timestamp': datetime.datetime.now().isoformat(),
                'python_version': platform.python_version(),
                'platform': platform.platform(),
                'container_info': 'Running in containerized environment'
            }
            self.wfile.write(json.dumps(response, indent=2).encode())
            
        elif self.path == '/health':
            self.send_response(200)
            self.send_header('Content-type', 'application/json')
            self.end_headers()
            self.wfile.write(b'{"status": "healthy"}')
            
        else:
            self.send_response(404)
            self.end_headers()

if __name__ == '__main__':
    server = HTTPServer(('0.0.0.0', 8000), APIHandler)
    print("🚀 Python API Server starting...")
    print("📍 Listening on http://0.0.0.0:8000")
    print("🔗 Endpoints: / (info), /health (status)")
    server.serve_forever()
'''

    # Data analysis script
    analysis_script = '''#!/usr/bin/env python3
import json
import statistics
import datetime

def analyze_data():
    # Sample employee data
    employees = [
        {'name': 'Alice', 'age': 25, 'salary': 50000, 'department': 'IT'},
        {'name': 'Bob', 'age': 30, 'salary': 75000, 'department': 'Finance'},
        {'name': 'Charlie', 'age': 35, 'salary': 85000, 'department': 'IT'},
        {'name': 'Diana', 'age': 28, 'salary': 60000, 'department': 'HR'},
        {'name': 'Eve', 'age': 32, 'salary': 90000, 'department': 'Finance'}
    ]
    
    print("📊 Employee Data Analysis (Container Edition)")
    print("=" * 50)
    
    # Basic statistics
    salaries = [emp['salary'] for emp in employees]
    ages = [emp['age'] for emp in employees]
    
    print(f"👥 Total Employees: {len(employees)}")
    print(f"💰 Average Salary: ${statistics.mean(salaries):,.2f}")
    print(f"📈 Salary Range: ${min(salaries):,} - ${max(salaries):,}")
    print(f"👤 Average Age: {statistics.mean(ages):.1f} years")
    
    # Department breakdown
    departments = {}
    for emp in employees:
        dept = emp['department']
        if dept not in departments:
            departments[dept] = []
        departments[dept].append(emp['salary'])
    
    print(f"\\n🏢 Department Analysis:")
    for dept, salaries in departments.items():
        avg_salary = statistics.mean(salaries)
        print(f"  {dept}: {len(salaries)} employees, avg ${avg_salary:,.2f}")
    
    # Save results
    results = {
        'analysis_date': datetime.datetime.now().isoformat(),
        'total_employees': len(employees),
        'average_salary': statistics.mean(salaries),
        'department_stats': {
            dept: {
                'count': len(sal),
                'average_salary': statistics.mean(sal)
            } for dept, sal in departments.items()
        }
    }
    
    print(f"\\n✅ Analysis complete!")
    return results

if __name__ == "__main__":
    results = analyze_data()
    print(f"\\n💾 Results: {json.dumps(results, indent=2)}")
'''

    return {
        'calculator.py': calculator_script,
        'api_server.py': api_script,
        'data_analysis.py': analysis_script
    }

# Generate and display scripts
scripts = generate_python_script()

print("📝 Generated Python Scripts for Container Testing:")
print("="*60)

for filename, content in scripts.items():
    print(f"\\n🐍 {filename}:")
    print(f"📏 Lines: {len(content.split(chr(10)))}")
    print(f"🎯 Purpose: ", end="")
    
    if 'calculator' in filename:
        print("Command-line calculator with error handling")
        print("💡 Usage: podman run --rm -v $(pwd):/app -w /app python:3.11 python calculator.py 10 + 5")
    elif 'api' in filename:
        print("Simple HTTP API server with JSON responses")
        print("💡 Usage: podman run -d -p 8000:8000 -v $(pwd):/app -w /app python:3.11 python api_server.py")
    elif 'analysis' in filename:
        print("Data analysis script with statistics")
        print("💡 Usage: podman run --rm -v $(pwd):/app -w /app python:3.11 python data_analysis.py")

print(f"\\n🚀 Container Commands Summary:")
print("```bash")
print("# Run calculator across different Python versions:")
print("podman run --rm -v $(pwd):/app -w /app python:3.9 python calculator.py 15 / 3")
print("podman run --rm -v $(pwd):/app -w /app python:3.11 python calculator.py 15 / 3")
print("podman run --rm -v $(pwd):/app -w /app python:3.12 python calculator.py 15 / 3")
print("")
print("# Run API server (accessible at localhost:8000):")
print("podman run -d --name api -p 8000:8000 -v $(pwd):/app -w /app python:3.11 python api_server.py")
print("")
print("# Run data analysis:")
print("podman run --rm -v $(pwd):/app -w /app python:3.11 python data_analysis.py")
print("```")

print(f"\\n💡 Key Benefits:")
print("✅ No Python installation needed on host")
print("✅ Test multiple Python versions instantly") 
print("✅ Isolated dependencies per project")
print("✅ Reproducible development environment")
````

## 🎯 Hands-On Exploration

````python
# Linux Distribution Container Explorer

def linux_distro_comparison():
    """Compare different Linux distributions available as containers"""
    
    distributions = {
        'ubuntu': {
            'description': 'Popular Debian-based distribution',
            'package_manager': 'apt',
            'size': '~70MB',
            'use_case': 'Development, web services',
            'commands': [
                'apt update',
                'apt install -y curl wget git',
                'cat /etc/os-release'
            ]
        },
        'alpine': {
            'description': 'Security-focused, lightweight distribution',
            'package_manager': 'apk',
            'size': '~5MB',
            'use_case': 'Production containers, microservices',
            'commands': [
                'apk update',
                'apk add python3 nodejs npm',
                'cat /etc/os-release'
            ]
        },
        'centos': {
            'description': 'Enterprise-focused Red Hat based',
            'package_manager': 'yum/dnf',
            'size': '~200MB',
            'use_case': 'Enterprise applications, servers',
            'commands': [
                'yum update -y',
                'yum install -y httpd mysql',
                'cat /etc/os-release'
            ]
        },
        'debian': {
            'description': 'Stable, reliable base distribution',
            'package_manager': 'apt',
            'size': '~50MB',
            'use_case': 'Servers, stable applications',
            'commands': [
                'apt update',
                'apt install -y build-essential',
                'cat /etc/os-release'
            ]
        },
        'fedora': {
            'description': 'Cutting-edge Red Hat technology',
            'package_manager': 'dnf',
            'size': '~200MB',
            'use_case': 'Development, latest software',
            'commands': [
                'dnf update -y',
                'dnf install -y python3 nodejs',
                'cat /etc/os-release'
            ]
        }
    }
    
    print("🐧 Linux Distribution Container Comparison")
    print("="*70)
    
    for distro, info in distributions.items():
        print(f"\\n📦 {distro.upper()}")
        print(f"   📝 {info['description']}")
        print(f"   📊 Size: {info['size']}")
        print(f"   🎯 Use Case: {info['use_case']}")
        print(f"   📦 Package Manager: {info['package_manager']}")
        
        print(f"   🚀 Container Command:")
        print(f"   podman run -it {distro} {'bash' if distro != 'alpine' else 'sh'}")
        
        print(f"   💻 Inside container commands:")
        for cmd in info['commands']:
            print(f"      {cmd}")
    
    print(f"\\n🎯 Quick Comparison Guide:")
    print("┌─────────────┬───────────┬─────────────┬─────────────────────┐")
    print("│ Distribution│    Size   │   Manager   │      Best For       │")
    print("├─────────────┼───────────┼─────────────┼─────────────────────┤")
    print("│ Alpine      │    5MB    │     apk     │ Production/Micro    │")
    print("│ Debian      │   50MB    │     apt     │ Stable servers      │")
    print("│ Ubuntu      │   70MB    │     apt     │ Development         │")
    print("│ CentOS      │  200MB    │   yum/dnf   │ Enterprise          │")
    print("│ Fedora      │  200MB    │     dnf     │ Latest tech         │")
    print("└─────────────┴───────────┴─────────────┴─────────────────────┘")
    
    return distributions

# Generate system information script for testing
def generate_system_info_script():
    """Generate a Python script to analyze container environment"""
    
    script = '''#!/usr/bin/env python3
import platform
import subprocess
import sys
import os

def get_system_info():
    """Gather comprehensive system information"""
    info = {
        'python': {
            'version': platform.python_version(),
            'implementation': platform.python_implementation(),
            'compiler': platform.python_compiler()
        },
        'system': {
            'os': platform.system(),
            'architecture': platform.machine(),
            'processor': platform.processor(),
            'platform': platform.platform()
        },
        'container': {
            'hostname': os.uname().nodename,
            'current_dir': os.getcwd(),
            'user_id': os.getuid() if hasattr(os, 'getuid') else 'N/A',
            'environment': dict(os.environ)
        }
    }
    return info

def detect_package_managers():
    """Detect available package managers"""
    managers = {
        'apt': ['apt', '--version'],
        'apk': ['apk', '--version'], 
        'yum': ['yum', '--version'],
        'dnf': ['dnf', '--version'],
        'pacman': ['pacman', '--version']
    }
    
    available = {}
    for pm, cmd in managers.items():
        try:
            result = subprocess.run(cmd, capture_output=True, text=True, timeout=5)
            if result.returncode == 0:
                version = result.stdout.split('\\n')[0]
                available[pm] = version
        except:
            pass
    
    return available

def get_memory_info():
    """Get memory information if available"""
    try:
        with open('/proc/meminfo', 'r') as f:
            lines = f.readlines()
            for line in lines:
                if 'MemTotal' in line:
                    mem_kb = int(line.split()[1])
                    return f"{mem_kb // 1024} MB"
    except:
        pass
    return "Unknown"

def main():
    print("🐧 Container System Analysis")
    print("=" * 50)
    
    info = get_system_info()
    
    print(f"🐍 Python Information:")
    print(f"   Version: {info['python']['version']}")
    print(f"   Implementation: {info['python']['implementation']}")
    
    print(f"\\n🖥️  System Information:")
    print(f"   OS: {info['system']['os']}")
    print(f"   Platform: {info['system']['platform']}")
    print(f"   Architecture: {info['system']['architecture']}")
    
    print(f"\\n📦 Package Managers:")
    managers = detect_package_managers()
    if managers:
        for pm, version in managers.items():
            print(f"   ✅ {pm}: {version}")
    else:
        print("   ❌ No standard package managers detected")
    
    print(f"\\n💾 Memory: {get_memory_info()}")
    print(f"🏠 Hostname: {info['container']['hostname']}")
    print(f"👤 User ID: {info['container']['user_id']}")
    
    # Check if we're in a container
    in_container = any([
        os.path.exists('/.dockerenv'),
        'container' in os.environ.get('TERM', ''),
        info['container']['hostname'].startswith(('docker-', 'podman-'))
    ])
    
    print(f"\\n🐋 Container Status: {'✅ Running in container' if in_container else '❌ Not in container'}")

if __name__ == "__main__":
    main()
'''
    
    return script

# Run the comparison
distros = linux_distro_comparison()

print(f"\\n\\n📝 Generated System Analysis Script:")
script = generate_system_info_script()
print(f"📏 Script length: {len(script.split(chr(10)))} lines")

print(f"\\n🚀 Usage Examples:")
print("```bash")
print("# Run system analysis on different distributions:")
print("podman run --rm -v $(pwd):/app -w /app ubuntu python3 /app/sysinfo.py")
print("podman run --rm -v $(pwd):/app -w /app alpine python3 /app/sysinfo.py") 
print("podman run --rm -v $(pwd):/app -w /app centos python3 /app/sysinfo.py")
print("")
print("# Interactive exploration:")
print("podman run -it ubuntu bash")
print("podman run -it alpine sh")
print("podman run -it fedora bash")
print("```")

print(f"\\n💡 Pro Tips:")
print("✅ Use Alpine for production (smallest size)")
print("✅ Use Ubuntu for development (most packages)")
print("✅ Use CentOS for enterprise (stable, supported)")
print("✅ Container vs VM: Same kernel, different userspace")
````

# 5️⃣ Networking & Multi-Container Applications

## 🌐 Container Networking Concepts

### Network Types
- **Bridge** (default): Isolated network for containers
- **Host**: Share host's network stack
- **None**: No network access

### Custom Networks
```bash
# Create custom network
podman network create myapp-network

# Run containers on same network
podman run -d --name database --network myapp-network postgres
podman run -d --name backend --network myapp-network myapi
podman run -d --name frontend --network myapp-network -p 80:80 nginx

# Containers can communicate using hostnames!
# backend connects to: postgresql://database:5432
```

## 🎯 Multi-Service Application Example

````python
# Multi-Container Application Architecture

def design_microservices_app():
    """Design a complete multi-container application"""
    
    architecture = {
        'frontend': {
            'image': 'nginx',
            'purpose': 'Web server serving static files',
            'ports': ['80:80'],
            'depends_on': ['backend'],
            'volumes': ['./frontend:/usr/share/nginx/html:ro'],
            'environment': []
        },
        'backend': {
            'image': 'python:3.11',
            'purpose': 'REST API server',
            'ports': ['5000:5000'],
            'depends_on': ['database', 'redis'],
            'volumes': ['./backend:/app'],
            'environment': [
                'DATABASE_URL=postgresql://database:5432/myapp',
                'REDIS_URL=redis://redis:6379',
                'FLASK_ENV=production'
            ]
        },
        'database': {
            'image': 'postgres:15-alpine',
            'purpose': 'PostgreSQL database',
            'ports': ['5432:5432'],
            'depends_on': [],
            'volumes': ['db-data:/var/lib/postgresql/data'],
            'environment': [
                'POSTGRES_PASSWORD=secret',
                'POSTGRES_DB=myapp',
                'POSTGRES_USER=admin'
            ]
        },
        'redis': {
            'image': 'redis:7-alpine',
            'purpose': 'In-memory cache and session store',
            'ports': ['6379:6379'],
            'depends_on': [],
            'volumes': ['redis-data:/data'],
            'environment': []
        }
    }
    
    print("🏗️  Multi-Container Application Architecture")
    print("="*60)
    
    print("📊 Service Overview:")
    for service, config in architecture.items():
        print(f"\\n🔧 {service.upper()}")
        print(f"   📦 Image: {config['image']}")
        print(f"   🎯 Purpose: {config['purpose']}")
        if config['ports']:
            print(f"   🔌 Ports: {', '.join(config['ports'])}")
        if config['depends_on']:
            print(f"   🔗 Depends on: {', '.join(config['depends_on'])}")
    
    # Generate deployment steps
    print(f"\\n🚀 Deployment Steps:")
    print("```bash")
    print("# 1. Create network")
    print("podman network create myapp-network")
    print("")
    
    # Order services by dependencies
    deployment_order = ['redis', 'database', 'backend', 'frontend']
    
    for service in deployment_order:
        config = architecture[service]
        cmd_parts = [f"podman run -d --name {service} --network myapp-network"]
        
        # Add ports
        for port in config['ports']:
            cmd_parts.append(f"-p {port}")
        
        # Add environment variables
        for env in config['environment']:
            cmd_parts.append(f"-e '{env}'")
        
        # Add volumes
        for vol in config['volumes']:
            cmd_parts.append(f"-v {vol}")
        
        cmd_parts.append(config['image'])
        
        if service == 'backend':
            cmd_parts.append("sh -c 'cd /app && python app.py'")
        
        command = " \\\\\n  ".join(cmd_parts)
        print(f"# {deployment_order.index(service) + 2}. Deploy {service}")
        print(command)
        print("")
    
    print("```")
    
    # Generate application flow
    print(f"\\n🔄 Application Flow:")
    print("┌─────────────┐    ┌─────────────┐    ┌─────────────┐    ┌─────────────┐")
    print("│   Browser   │───▶│  Frontend   │───▶│   Backend   │───▶│  Database   │")
    print("│ (Port 80)   │    │   (Nginx)   │    │  (Python)   │    │(PostgreSQL) │")
    print("└─────────────┘    └─────────────┘    └─────────────┘    └─────────────┘")
    print("                                             │                      ")
    print("                                             ▼                      ")
    print("                                    ┌─────────────┐                 ")
    print("                                    │    Redis    │                 ")
    print("                                    │   (Cache)   │                 ")
    print("                                    └─────────────┘                 ")
    
    # Generate sample application code
    print(f"\\n📝 Sample Backend Code (Python Flask):")
    sample_backend = '''
from flask import Flask, jsonify, request
import psycopg2
import redis
import os

app = Flask(__name__)

# Database connection
def get_db():
    return psycopg2.connect(
        host='database',
        database=os.getenv('POSTGRES_DB', 'myapp'),
        user=os.getenv('POSTGRES_USER', 'admin'),
        password=os.getenv('POSTGRES_PASSWORD', 'secret')
    )

# Redis connection  
redis_client = redis.Redis(host='redis', port=6379, decode_responses=True)

@app.route('/api/health')
def health():
    return jsonify({'status': 'healthy', 'service': 'backend'})

@app.route('/api/users')
def get_users():
    # Check cache first
    cached = redis_client.get('users')
    if cached:
        return cached
    
    # Query database
    conn = get_db()
    cur = conn.cursor()
    cur.execute("SELECT * FROM users LIMIT 10")
    users = cur.fetchall()
    
    # Cache result
    redis_client.setex('users', 300, jsonify(users))
    return jsonify(users)

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
'''
    
    print("```python")
    print(sample_backend.strip())
    print("```")
    
    return architecture

# Run the architecture design
app_arch = design_microservices_app()

print(f"\\n💡 Key Networking Concepts:")
print("✅ Container name resolution (DNS)")
print("✅ Network isolation and security") 
print("✅ Port mapping for external access")
print("✅ Service discovery within network")
print("✅ Environment variable configuration")

print(f"\\n🔧 Troubleshooting Tips:")
print("• Check network connectivity: podman exec backend ping database")
print("• Verify environment variables: podman exec backend printenv")
print("• Check logs: podman logs backend")
print("• Monitor network: podman network inspect myapp-network")
````

````python
# Container Debugging Toolkit

def container_troubleshooting_guide():
    """Comprehensive container debugging scenarios"""
    
    scenarios = {
        'container_wont_start': {
            'problem': "Container exits immediately after starting",
            'symptoms': [
                "Container shows 'Exited' status immediately",
                "No application output in logs",
                "Exit code is non-zero"
            ],
            'diagnosis': [
                "podman logs containerName",
                "podman inspect containerName --format '{{.State.ExitCode}}'",
                "podman run -it imageName bash  # Test image directly"
            ],
            'solutions': [
                "Check if the main process stays running (not one-shot)",
                "Verify command syntax in Dockerfile/run command",
                "Check if required files/dependencies exist",
                "Ensure proper permissions on executable files"
            ]
        },
        'network_connectivity': {
            'problem': "Container can't connect to other services",
            'symptoms': [
                "Connection refused errors",
                "DNS resolution failures",
                "Timeouts when connecting to databases/APIs"
            ],
            'diagnosis': [
                "podman exec container ping otherService",
                "podman exec container nslookup otherService", 
                "podman network inspect networkName",
                "podman port container  # Check port mappings"
            ],
            'solutions': [
                "Ensure containers are on the same network",
                "Use container names for inter-service communication",
                "Check firewall rules and port mappings",
                "Verify service is listening on 0.0.0.0, not 127.0.0.1"
            ]
        },
        'performance_issues': {
            'problem': "Container running slowly or using too many resources",
            'symptoms': [
                "High CPU or memory usage",
                "Application responding slowly",
                "Host system becoming unresponsive"
            ],
            'diagnosis': [
                "podman stats container",
                "podman exec container top",
                "podman exec container df -h  # Check disk usage",
                "podman system df  # Check Podman disk usage"
            ],
            'solutions': [
                "Set resource limits: --memory=512m --cpus=1.5",
                "Optimize application code and queries",
                "Use multi-stage builds to reduce image size",
                "Monitor and clean up unused volumes/images"
            ]
        },
        'volume_permission': {
            'problem': "Permission denied when accessing mounted volumes",
            'symptoms': [
                "Permission denied errors in logs",
                "Cannot read/write to mounted directories",
                "Application fails to access data files"
            ],
            'diagnosis': [
                "podman exec container ls -la /mountpoint",
                "podman exec container id  # Check user ID",
                "ls -la hostPath  # Check host permissions",
                "podman inspect container --format '{{.Config.User}}'"
            ],
            'solutions': [
                "Change host directory permissions: chmod 755 hostPath",
                "Run container as specific user: --user 1000:1000",
                "Use SELinux labels: :Z or :z in volume mount",
                "Change ownership: chown -R user:group hostPath"
            ]
        }
    }
    
    print("🔧 Container Troubleshooting Guide")
    print("="*60)
    
    for scenario, details in scenarios.items():
        print(f"\\n🚨 Scenario: {details['problem']}")
        print("-" * 40)
        
        print("📋 Symptoms:")
        for symptom in details['symptoms']:
            print(f"   • {symptom}")
        
        print("\\n🔍 Diagnosis Commands:")
        for cmd in details['diagnosis']:
            print(f"   $ {cmd}")
        
        print("\\n💡 Solutions:")
        for solution in details['solutions']:
            print(f"   ✅ {solution}")
    
    return scenarios

def generate_debugging_script():
    """Generate a comprehensive container health check script"""
    
    script = '''#!/bin/bash
# Container Health Check Script

CONTAINER_NAME=${1:-"mycontainer"}

echo "🔍 Container Health Check: $CONTAINER_NAME"
echo "=================================="

# Check if container exists
if ! podman container exists "$CONTAINER_NAME"; then
    echo "❌ Container '$CONTAINER_NAME' does not exist"
    echo "Available containers:"
    podman ps -a --format "{{.Names}}"
    exit 1
fi

echo "✅ Container exists"

# Check container status
STATUS=$(podman inspect "$CONTAINER_NAME" --format '{{.State.Status}}')
echo "📊 Status: $STATUS"

# If running, show detailed info
if [ "$STATUS" = "running" ]; then
    echo "\\n🔄 Runtime Information:"
    echo "Started: $(podman inspect "$CONTAINER_NAME" --format '{{.State.StartedAt}}')"
    echo "PID: $(podman inspect "$CONTAINER_NAME" --format '{{.State.Pid}}')"
    
    echo "\\n💾 Resource Usage:"
    podman stats "$CONTAINER_NAME" --no-stream
    
    echo "\\n🌐 Network Information:"
    podman inspect "$CONTAINER_NAME" --format '{{range .NetworkSettings.Networks}}IP: {{.IPAddress}}, Gateway: {{.Gateway}}{{end}}'
    
    echo "\\n🔌 Port Mappings:"
    podman port "$CONTAINER_NAME"
    
    echo "\\n📁 Volume Mounts:"
    podman inspect "$CONTAINER_NAME" --format '{{range .Mounts}}{{.Source}} -> {{.Destination}} ({{.Type}}){{end}}'
    
    echo "\\n🔧 Environment Variables:"
    podman exec "$CONTAINER_NAME" printenv | head -10
    
    echo "\\n📋 Running Processes:"
    podman exec "$CONTAINER_NAME" ps aux 2>/dev/null || echo "Cannot access process list"
    
else
    echo "\\n❌ Container is not running"
    echo "Exit Code: $(podman inspect "$CONTAINER_NAME" --format '{{.State.ExitCode}}')"
    echo "\\n📄 Last 20 log lines:"
    podman logs --tail 20 "$CONTAINER_NAME"
fi

echo "\\n✅ Health check complete"
'''
    
    return script

# Run troubleshooting guide
scenarios = container_troubleshooting_guide()

print(f"\\n\\n📝 Generated Health Check Script:")
script = generate_debugging_script()
lines = len(script.split('\\n'))
print(f"📏 Script: {lines} lines of bash code")

print(f"\\n🚀 Usage Examples:")
print("```bash")
print("# Save the health check script")
print("cat > container_health_check.sh << 'EOF'")
print(script.strip())
print("EOF")
print("")
print("# Make it executable and run")
print("chmod +x container_health_check.sh")
print("./container_health_check.sh mycontainer")
print("")
print("# Quick debugging commands:")
print("podman logs -f mycontainer          # Follow logs in real-time")
print("podman exec -it mycontainer bash    # Interactive shell") 
print("podman inspect mycontainer | jq     # Pretty JSON output")
print("podman stats --no-stream            # Current resource usage")
print("```")

print(f"\\n🎯 Debugging Best Practices:")
print("✅ Always check logs first: `podman logs container`")
print("✅ Use health checks in production containers")
print("✅ Set resource limits to prevent resource exhaustion")
print("✅ Test container images locally before deployment")
print("✅ Use proper logging levels (DEBUG in dev, INFO in prod)")
print("✅ Monitor disk space - containers can fill up host storage")

print(f"\\n🔧 Emergency Commands:")
print("• Stop all containers: `podman stop $(podman ps -q)`")
print("• Remove all stopped containers: `podman container prune`") 
print("• Clean up disk space: `podman system prune --volumes`")
print("• Kill unresponsive container: `podman kill containerName`")
````

### Complete System Cleanup
```bash

podman system prune

podman system prune --volumes --all
```

````python
# Container Maintenance & Cleanup Automation

def generate_cleanup_strategy():
    """Generate automated cleanup strategies"""
    
    strategies = {
        'development': {
            'description': 'Aggressive cleanup for development environments',
            'frequency': 'Daily',
            'commands': [
                'podman container prune -f',
                'podman image prune -f',
                'podman volume prune -f',
                'podman network prune -f'
            ],
            'safety_level': 'Low - removes all unused resources'
        },
        'staging': {
            'description': 'Moderate cleanup for staging environments',
            'frequency': 'Weekly',
            'commands': [
                'podman container prune --filter "until=48h" -f',
                'podman image prune --filter "until=168h" -f',  # 1 week
                'podman volume prune --filter "label!=keep" -f'
            ],
            'safety_level': 'Medium - preserves recent resources'
        },
        'production': {
            'description': 'Conservative cleanup for production',
            'frequency': 'Monthly',
            'commands': [
                'podman container prune --filter "until=720h" -f',  # 30 days
                'podman image prune --filter "dangling=true" -f',
                # Only remove clearly unused volumes
                'podman volume ls -q --filter "dangling=true" | xargs -r podman volume rm'
            ],
            'safety_level': 'High - only removes clearly unused resources'
        }
    }
    
    print("🧹 Container Cleanup Strategies")
    print("="*50)
    
    for env, strategy in strategies.items():
        print(f"\\n🏷️  {env.upper()} Environment")
        print(f"📝 {strategy['description']}")
        print(f"⏰ Frequency: {strategy['frequency']}")
        print(f"🛡️  Safety: {strategy['safety_level']}")
        print(f"💻 Commands:")
        for cmd in strategy['commands']:
            print(f"   {cmd}")
    
    return strategies

def simulate_disk_usage_check():
    """Simulate container disk usage analysis"""
    
    # Simulated data
    disk_usage = {
        'images': {
            'total_size': '4.2 GB',
            'count': 15,
            'reclaimable': '1.8 GB',
            'details': [
                {'name': 'python:3.11', 'size': '900 MB', 'used': True},
                {'name': 'nginx:latest', 'size': '140 MB', 'used': True}, 
                {'name': 'redis:alpine', 'size': '30 MB', 'used': False},
                {'name': 'postgres:15', 'size': '380 MB', 'used': True},
                {'name': 'ubuntu:latest', 'size': '70 MB', 'used': False}
            ]
        },
        'containers': {
            'total_size': '850 MB', 
            'count': 8,
            'running': 3,
            'stopped': 5,
            'reclaimable': '450 MB'
        },
        'volumes': {
            'total_size': '2.1 GB',
            'count': 6,
            'reclaimable': '800 MB',
            'details': [
                {'name': 'db_data', 'size': '1.2 GB', 'used': True},
                {'name': 'cache_data', 'size': '300 MB', 'used': False},
                {'name': 'logs_vol', 'size': '600 MB', 'used': True}
            ]
        }
    }
    
    print("📊 Disk Usage Analysis")
    print("="*40)
    
    print(f"🖼️  Images: {disk_usage['images']['total_size']} ({disk_usage['images']['count']} images)")
    print(f"   💾 Reclaimable: {disk_usage['images']['reclaimable']}")
    
    print(f"\\n🔧 Containers: {disk_usage['containers']['total_size']} ({disk_usage['containers']['count']} containers)")
    print(f"   🟢 Running: {disk_usage['containers']['running']}")
    print(f"   🔴 Stopped: {disk_usage['containers']['stopped']}")
    print(f"   💾 Reclaimable: {disk_usage['containers']['reclaimable']}")
    
    print(f"\\n💾 Volumes: {disk_usage['volumes']['total_size']} ({disk_usage['volumes']['count']} volumes)")
    print(f"   💾 Reclaimable: {disk_usage['volumes']['reclaimable']}")
    
    print(f"\\n📋 Detailed Breakdown:")
    print("Images:")
    for img in disk_usage['images']['details']:
        status = "🟢 Used" if img['used'] else "🔴 Unused"
        print(f"   {img['name']:<20} {img['size']:<10} {status}")
    
    print("\\nVolumes:")
    for vol in disk_usage['volumes']['details']:
        status = "🟢 Used" if vol['used'] else "🔴 Unused" 
        print(f"   {vol['name']:<20} {vol['size']:<10} {status}")
    
    # Calculate potential savings
    total_reclaimable = 1.8 + 0.45 + 0.8  # GB
    print(f"\\n💰 Total Reclaimable Space: ~{total_reclaimable:.1f} GB")
    
    return disk_usage

def generate_maintenance_script():
    """Generate automated maintenance script"""
    
    script = '''#!/bin/bash
# Container Maintenance Script
# Run this script periodically to keep your container environment clean

set -e  # Exit on error

ENVIRONMENT=${1:-"development"}
DRY_RUN=${2:-"false"}

echo "🧹 Container Maintenance Script"
echo "Environment: $ENVIRONMENT"
echo "Dry Run: $DRY_RUN"
echo "================================"

# Function to execute or simulate command
run_cmd() {
    local cmd="$1"
    local desc="$2"
    
    echo "🔄 $desc"
    if [ "$DRY_RUN" = "true" ]; then
        echo "   [DRY RUN] Would execute: $cmd"
    else
        echo "   Executing: $cmd"
        eval "$cmd"
    fi
    echo ""
}

# Check disk usage before cleanup
echo "📊 Disk usage before cleanup:"
podman system df

case "$ENVIRONMENT" in
    "development")
        echo "🔧 Running development cleanup (aggressive)..."
        run_cmd "podman container prune -f" "Remove stopped containers"
        run_cmd "podman image prune -f" "Remove unused images" 
        run_cmd "podman volume prune -f" "Remove unused volumes"
        run_cmd "podman network prune -f" "Remove unused networks"
        ;;
        
    "staging")
        echo "⚖️  Running staging cleanup (moderate)..."
        run_cmd "podman container prune --filter 'until=48h' -f" "Remove containers older than 2 days"
        run_cmd "podman image prune --filter 'until=168h' -f" "Remove images older than 1 week"
        run_cmd "podman volume prune --filter 'label!=keep' -f" "Remove unprotected volumes"
        ;;
        
    "production") 
        echo "🛡️  Running production cleanup (conservative)..."
        run_cmd "podman container prune --filter 'until=720h' -f" "Remove containers older than 30 days"
        run_cmd "podman image prune --filter 'dangling=true' -f" "Remove dangling images only"
        echo "ℹ️  Skipping volume cleanup in production (manual review required)"
        ;;
        
    *)
        echo "❌ Unknown environment: $ENVIRONMENT"
        echo "Valid environments: development, staging, production"
        exit 1
        ;;
esac

# Check disk usage after cleanup
echo "📊 Disk usage after cleanup:"
podman system df

echo "✅ Maintenance complete!"
'''
    
    return script

# Run the maintenance tools
print("🛠️  Container Maintenance Tools")
print("="*50)

strategies = generate_cleanup_strategy()
disk_usage = simulate_disk_usage_check()

print(f"\\n\\n📝 Generated Maintenance Script:")
script = generate_maintenance_script()
print(f"📏 Script: {len(script.split(chr(10)))} lines")

print(f"\\n🚀 Usage Examples:")
print("```bash")
print("# Save the maintenance script")
print("curl -o container_maintenance.sh << 'EOF'")
print("# [script content here]")
print("EOF")
print("")
print("# Make executable and run")
print("chmod +x container_maintenance.sh") 
print("")
print("# Different environments")
print("./container_maintenance.sh development     # Aggressive cleanup")
print("./container_maintenance.sh staging         # Moderate cleanup")
print("./container_maintenance.sh production      # Conservative cleanup")
print("")
print("# Dry run mode (shows what would be done)")
print("./container_maintenance.sh development true")
print("```")

print(f"\\n⚠️  Safety Guidelines:")
print("🔴 NEVER run aggressive cleanup on production")
print("🟡 Always test cleanup scripts in development first") 
print("🟢 Use dry run mode to preview changes")
print("💾 Backup important volumes before cleanup")
print("📅 Schedule regular maintenance (not just when disk is full)")

print(f"\\n📊 Monitoring Commands:")
print("• Check disk usage: `podman system df`")
print("• List all resources: `podman system info`")
print("• Monitor in real-time: `watch -n 5 podman system df`")
````
